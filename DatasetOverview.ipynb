```python
import pandas as pd
import numpy as np
from windrose import WindroseAxes
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv').drop(['Time - Unix Format'], axis=1)
```

```python
df
```

```python
ax = WindroseAxes.from_ax()
ax.contourf(df['Wind Degree [°]'], df['Wind Speed [m/s]'], normed=True, bins=np.arange(0, 12, 1), cmap=cm.winter_r)
ax.set_legend(loc='lower right', bbox_to_anchor=(1.35, 0.2), prop={'size': 14})
ax.set_ylabel
ax.set_xlabel
fig = plt.gcf()
fig.set_size_inches(10, 6)
```
